# Question5:submit the code that calculates the following batting rank score for the 1981 Reds (CIN)

    = G(H/AB)+(RBI)/R

where G,H,AB,RBI,RG,H,AB,RBI,R are games, hits, at bats, runs batted in and runs respectively.

Your answer should ONLY include the batting for the 1981 Cincinnati Reds batting roster.
where G,H,AB,RBI,RG,H,AB,RBI,R are games, hits, at bats, runs batted in and runs respectively.

Your answer should ONLY include the batting for the 1981 Cincinnati Reds batting roster.

```python
#StudentName: Venkateshwara Reddy Kanumuri, StudentID:999992999, Section:029
#Program for loading csv file to pandas dataframe
#Program for querying CIN->1981 players records and converting that to JSON object.
#Program to caluculate players score and pick out the top 4 players.
import pandas as pd
import json
from pprint import pprint
import operator
from itertools import islice

df1=pd.read_csv("https://raw.githubusercontent.com/chadwickbureau/baseballdatabank/master/core/Batting.csv")

df2 = df1[(df1.yearID == 1981) & (df1.teamID == 'CIN')]

df3 = df2[['playerID','G','H','AB','RBI','R']]


df3.reset_index().to_json(path_or_buf="cin.json",orient='records')

with open('cin.json') as data_file:    
    data = json.load(data_file)


playerScore = {}

for record in data:
	pID=record['playerID']
	pGames=record['G']
	pHits=record['H']
	pAtBats=record['AB']
	pRBI=record['RBI']
	pRuns=record['R']
	
	if pRuns == 0:
		continue
	if pAtBats == 0:
		continue
	sqrtRuns=pRuns**(1/2);
	result=pGames*(pHits/pAtBats) + (pRBI/sqrtRuns)
	playerScore[pID]=result


sorted_score = sorted(playerScore.items(), key=operator.itemgetter(1), reverse=True)
print("Players with their score list:")
for x in sorted_score:
    player,score=x
    print("Player Name:"+player+"  Score:"+ str(score))

```

# Question6: for the 1981 data submit the top 4 batter IDs by their batting rank score, , previously calculated.

```python
#StudentName: Venkateshwara Reddy Kanumuri, StudentID:999992999, Section:029
#Program for loading csv file to pandas dataframe
#Program for querying CIN->1981 players records and converting that to JSON object.
#Program to caluculate players score and pick out the top 4 players.
import pandas as pd
import json
from pprint import pprint
import operator
from itertools import islice

df1=pd.read_csv("https://raw.githubusercontent.com/chadwickbureau/baseballdatabank/master/core/Batting.csv")

df2 = df1[(df1.yearID == 1981) & (df1.teamID == 'CIN')]

df3 = df2[['playerID','G','H','AB','RBI','R']]


df3.reset_index().to_json(path_or_buf="cin.json",orient='records')

with open('cin.json') as data_file:    
    data = json.load(data_file)


playerScore = {}

for record in data:
	pID=record['playerID']
	pGames=record['G']
	pHits=record['H']
	pAtBats=record['AB']
	pRBI=record['RBI']
	pRuns=record['R']
	
	if pRuns == 0:
		continue
	if pAtBats == 0:
		continue
	sqrtRuns=pRuns**(1/2);
	result=pGames*(pHits/pAtBats) + (pRBI/sqrtRuns)
	playerScore[pID]=result


sorted_score = sorted(playerScore.items(), key=operator.itemgetter(1), reverse=True)

print("\nTop 4 players:")
i=0;
for x in sorted_score:
	i=i+1;
	if i>4:
		break
	else:
		player,score=x
		print("Player Name:"+player+"  Score:"+ str(score) +" Rank:"+ str(i))


```
